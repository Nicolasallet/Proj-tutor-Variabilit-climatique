```r
library('dplyr')
library("zoo")
directory = '/media/nicolas/Elements/data/'
time_past = 1970
time_pre = 2010
STAID='000013'
time=time_pre
path_to_tempmean = paste(directory,'TG/TG_STAID',STAID,'.txt',sep='')
```

```r
openfile <- function(path_to_data){
    Data <- read.table(path_to_data,skip = 20,header=TRUE,sep=',')
    Data <- filter(Data,Data[,5]==0)               
    Data <- Data[,-c(1,2,5)]
    Data$DATE <- strptime(Data$DATE, format = "%Y%m%d")
 
    Data <- filter ( Data, Data$DATE$year > time - 1905 & Data$DATE$year <= time - 1900)
    return(Data)
}
```

```r
data <- openfile (path_to_tempmean)
data
```
