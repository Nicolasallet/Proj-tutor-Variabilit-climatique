```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import cartopy
import cartopy.crs as crs
import cartopy.feature as cf

import sys
from itertools import combinations
from datetime import datetime
import random
import warnings

warnings.filterwarnings("ignore")

print("Python Version : ", sys.version)
print("Cartopy Version : ", cartopy.__version__)

%matplotlib inline
```

```python
df = pd.read_csv('C:/Users/lilia/Projet tutor√©/megatableau_dist_anglesfinal80_2020.csv')
df.head()
```

```python
fig = plt.figure(figsize=(20,16))
ax = fig.add_subplot(1,1,1, projection=crs.Robinson())
ax.coastlines()
ax.gridlines()
plt.scatter(x=df.LON_x, y=df.LAT_x,
            color="red",
            s=5,
            alpha=1,
            transform=crs.PlateCarree())

ax.add_feature(cf.LAND) 
ax.add_feature(cf.BORDERS) 
ax.add_feature(cf.COASTLINE)
```

```python
fig = plt.figure(figsize=(20,16))
ax = fig.add_subplot(1,1,1, projection=crs.Robinson())
ax.coastlines()
ax.gridlines()
plt.scatter(x=df.LON_y, y=df.LAT_y,
            color="blue",
            s=5,
            alpha=1,
            transform=crs.PlateCarree())

ax.add_feature(cf.LAND) 
ax.add_feature(cf.BORDERS) 
ax.add_feature(cf.COASTLINE)
```

```python
fig = plt.figure(figsize=(20,16))
ax = fig.add_subplot(1,1,1, projection=crs.Robinson())
ax.coastlines()
ax.gridlines()

for idx in df.index:

    lat1, lon1 = df.loc[idx]["LAT_x"], df.loc[idx]["LON_x"]
    lat2, lon2 = df.loc[idx]["LAT_y"], df.loc[idx]["LON_y"]
    
    plt.plot([lon1, lon2], [lat1, lat2],
             linewidth=0.1,
             color="black",
             transform=crs.PlateCarree()) 
    
    plt.scatter(x=df.LON_x, y=df.LAT_x,
            color="red",
            s=5,
            alpha=1,
            transform=crs.PlateCarree())

    plt.scatter(x=df.LON_y, y=df.LAT_y,
            color="blue",
            s=5,
            alpha=1,
            transform=crs.PlateCarree())

ax.add_feature(cf.LAND) 
ax.add_feature(cf.BORDERS) 
ax.add_feature(cf.COASTLINE)
```

```python
fig = plt.figure(figsize=(20,16))
ax = fig.add_subplot(1,1,1, projection=crs.Robinson())
ax.coastlines()
ax.gridlines()

for i in df.index :
    if df.Dist_lat[i] <= 0 : 
        plt.scatter(x=df.LON_x[i], y=df.LAT_x[i],
            color="red",
            s=[0.05*abs(df.Dist_lat[i])],
            alpha=1,
            transform=crs.PlateCarree())
    else : 
        plt.scatter(x=df.LON_x[i], y=df.LAT_x[i],
                color="blue",
                s=[0.05*abs(df.Dist_lat[i])],
                alpha=1,
                transform=crs.PlateCarree())
        
ax.add_feature(cf.LAND) 
ax.add_feature(cf.BORDERS) 
ax.add_feature(cf.COASTLINE)
```

```python
fig = plt.figure(figsize=(20,16))
ax = fig.add_subplot(1,1,1, projection=crs.Robinson())

ax.coastlines()
ax.gridlines()

for i in df.index :
    if df.Dist_lon[i] <= 0 : 
        plt.scatter(x=df.LON_x[i], y=df.LAT_x[i],
            color="green",
            s=[0.05*abs(df.Dist_lon[i])],
            alpha=1,
            transform=crs.PlateCarree())
    else :
        plt.scatter(x=df.LON_x[i], y=df.LAT_x[i],
                color="orange",
                s=[0.05*abs(df.Dist_lon[i])],
                alpha=1,
                transform=crs.PlateCarree())
        
ax.add_feature(cf.LAND) 
ax.add_feature(cf.BORDERS) 
ax.add_feature(cf.COASTLINE)
```
