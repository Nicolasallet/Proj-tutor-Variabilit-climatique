```r
rad2deg <- function(rad) {(rad * 180)/(pi)}
deg2rad <- function(deg) {(deg * pi )/(180)}
```

```r
for (i in 1:length(megatableau$X)){
    
deltalon=megatableau[i,'LON_y']-megatableau[i,'LON_x']
deltalat=(megatableau[i,'LAT_y']-megatableau[i,'LAT_x'])
lony=deg2rad(megatableau[i,'LON_y'])
lonx=deg2rad(megatableau[i,'LON_x'])
laty=deg2rad(megatableau[i,'LAT_y'])
latx=deg2rad(megatableau[i,'LAT_x'])
dlo= abs(lony-lonx)
dLa= abs(laty-latx )
    
megatableau[i,'D_ALT']=megatableau[i,'ALT_y']-megatableau[i,'ALT_x']   
    #calcul des distances latitudinales et logitudinales 
megatableau[i,"Dist_lat"]<-deltalat*111 
megatableau[i,"Dist_lon"] <-deltalon*111*cos(latx)
    
#calcul de la distance séparant deux stations 
dist_angle_rad =acos( sin(latx)*sin(laty) +cos(latx)*cos(laty)*  cos(abs(lonx-lony)))
    dist= dist_angle_rad * 6378
    megatableau[i,"Dist"]<-dist
#calcul de l'orientation du trajet en degré vers le NE (O°-> vers le nord)
X = cos(laty)* sin(dlo)
Y = cos(latx)*sin(laty)- sin(latx)*cos(laty)* cos(dlo)

    angle=360-rad2deg(atan2(X,Y))
       
     megatableau[i,"angle"]<-angle
    
    }
# calcul de statistiques sur les données 
mean(megatableau$Dist_lat)
median(megatableau$Dist_lat)
megatableau

#write.table(megatableau,'C:/Users/Brune/Desktop/JUPYTER/VCE_ville_climatique/megatableau_dist_anglesfinal80_2020.csv',
          # row.names=FALSE,sep=',')
```
